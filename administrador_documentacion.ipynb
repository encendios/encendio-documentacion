Este manual proporciona una gua paso a paso para la administracin de cuentas de usuario en el panel de Keycloak, incluyendo la creacin y deshabilitacin de cuentas de administrador.

### Creacin de Cuentas para Administrador

Para asignar un rol de administrador a un usuario, sigue los pasos a continuacin:

1. **Acceso al Panel de Keycloak**  
   Primero, asegrate de tener las credenciales necesarias (usuario y contrasea) para acceder al panel de Keycloak. Una vez que tengas las credenciales, ingresa a la consola de administracin de Keycloak. [http://auth.146.83.216.195.nip.io/](http://auth.146.83.216.195.nip.io/)

   

2. **Navegacin en la Consola de Keycloak**  
   Una vez dentro de la consola de Keycloak, te encontrars en el dashboard principal.

  

3. **Bsqueda y Seleccin del Usuario**  
   Dirgete a la seccin de "Users" en el men lateral. En el campo de bsqueda, escribe el correo electrnico del usuario que deseas convertir en administrador y haz clic sobre su nombre para acceder a su perfil.

  

4. **Asignacin del Rol de Administrador**  
   En el perfil del usuario, selecciona la pestaa "Role Mappings". En la seccin "Client Roles", elige "dashboard" del men desplegable. Despus, selecciona "administradorCrud" de la lista de roles disponibles y haz clic en "Add selected" para asignar el rol al usuario.

  

5. **Actualizacin de los Privilegios del Usuario**  
   Una vez que el rol de administrador ha sido asignado, el usuario deber cerrar sesin y volver a ingresar para que los cambios surtan efecto.

### Deshabilitar Administrador

Si es necesario revocar los permisos de administrador de un usuario, sigue los siguientes pasos:

1. **Acceso al Perfil del Usuario**  
   Ingresa nuevamente al panel de Keycloak con las credenciales de administrador. Navega a la seccin "Users" y selecciona el usuario al que deseas deshabilitar como administrador.



2. **Remocin del Rol de Administrador**  
   En el perfil del usuario, ve a la pestaa "Role Mappings". En "Client Roles", selecciona "dashboard" del men desplegable. Despus, selecciona "administradorCrud" de la lista de roles asignados y haz clic en "Remove selected" para eliminar el rol de administrador del usuario.

 

Una vez que el rol ha sido removido, el usuario perder inmediatamente los privilegios de administrador.

---

### Notas Adicionales

- Asegrate de que el usuario cierre sesin y vuelva a ingresar despus de realizar cambios en sus roles para que los cambios sean efectivos.
- Solo los usuarios con el rol adecuado pueden asignar o revocar permisos de administrador en Keycloak.


## Panel de Administracin

El panel de administracin es una herramienta integral que te permite gestionar varios elementos relacionados con la plataforma Encendios. A travs de este panel, puedes administrar palabras, frases, imgenes, ligas, items, y elementos desactivados de manera eficiente y sencilla.



#### Elementos Administrables

1. **Palabras**:  
   En esta seccin, puedes realizar las siguientes acciones:
   - **Ver detalles de creacin de palabras**: Consulta informacin sobre cundo y quin cre una palabra especfica.
   - **ltima actualizacin**: Revisa la fecha de la ltima modificacin realizada a una palabra.
   - **Crear nuevas palabras**: Aade nuevas palabras al sistema.
   - **Editar palabras existentes**: Modifica el contenido, significado o atributos de palabras ya registradas.
   - **Ocultar palabras**: Marca palabras para que no sean visibles en la aplicacin, sin eliminarlas permanentemente del sistema.

   

2. **Frases**:  
   La seccin de frases te permite:
   - **Visualizar frases existentes**: Revisa todas las frases registradas en la plataforma.
   - **Fechas de creacin y actualizacin**: Consulta las fechas en las que se crearon y actualizaron las frases.
   - **Editar frases**: Realiza cambios en las frases, ya sea en el contenido o en sus atributos.
   - **Ocultar frases**: Similar a las palabras, puedes ocultar frases para que no sean visibles en la aplicacin sin eliminarlas.

   

3. **Imgenes**:  
   Desde aqu, puedes gestionar todas las imgenes del sistema:
   - **Visualizar imgenes**: Observa todas las imgenes disponibles en la plataforma.
   - **Crear nuevas imgenes**: Sube imgenes nuevas para su uso en la aplicacin.
   - **Editar imgenes**: Modifica atributos de las imgenes, como su descripcin o categora.
   - **Ver detalles de subida**: Consulta quin ha subido cada imagen y en qu categora se encuentran.
   - **Editar atributos**: Ajusta los atributos asociados a las imgenes, como la categora, etiquetas, etc.

   

4. **Ligas**:  
   Esta seccin est diseada para la gestin de ligas y rangos:
   - **Definir atributos de liga**: Configura los diferentes atributos necesarios para definir las ligas y los rangos dentro de ellas.
   - **Tamao de la liga**: Establece el tamao de cada liga, es decir, cuntos jugadores pueden estar en una liga determinada.
   - **Activar o desactivar ligas**: Habilita o inhabilita ligas segn las necesidades del sistema o del juego.
   - **Editar ligas**: Realiza ajustes en las configuraciones de las ligas ya existentes.

   

5. **Items**:  
   Administra todos los elementos relacionados con la gamificacin:
   - **Crear items**: Aade nuevos elementos de gamificacin al sistema.
   - **Editar items**: Modifica las caractersticas y atributos de los items ya existentes.
   - **Ocultar items**: Elimina temporalmente items del acceso pblico sin borrarlos definitivamente.

   

6. **Desactivados**:  
   Esta seccin permite visualizar y gestionar los elementos desactivados:
   - **Frases desactivadas**: Muestra todas las frases que han sido ocultadas.
   - **Palabras desactivadas**: Lista de palabras que estn desactivadas.
   - **Imgenes desactivadas**: Imgenes que no estn visibles actualmente en la aplicacin.
   - **Items desactivados**: Items que se encuentran desactivados en el sistema.
   - **Ligas desactivadas**: Muestra todas las ligas que han sido desactivadas.
   - **Reactivar elementos**: Desde esta seccin, puedes volver a activar cualquier elemento previamente desactivado, devolvindolo al estado activo en la aplicacin.

   

### Notas Importantes

- **Acceso restringido**: Asegrate de que solo los usuarios con los roles adecuados tengan acceso a las funcionalidades de administracin.
- **Cambios en tiempo real**: Los cambios realizados en el panel de administracin pueden surtir efecto inmediato, por lo que es importante revisar las modificaciones antes de confirmarlas.
- **Historial de cambios**: Mantn un registro de todas las modificaciones realizadas para facilitar auditoras y revisiones futuras.

Este panel es una herramienta poderosa que permite mantener la plataforma Encendios actualizada, organizada y en buen funcionamiento. Para ms detalles o instrucciones especficas sobre el uso del panel de administracin, consulta la documentacin tcnica o comuncate con el equipo de soporte.
